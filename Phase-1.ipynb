# Python – Overview

## What is Python?
- Python is a **high-level, interpreted programming language**
- Created by **Guido van Rossum**, released in **1991**
- Designed for **readability, simplicity, and fast development**

## Use Cases
- 🌐 Web development (Django, Flask, FastAPI)
- ⚙️ Software development & automation
- 📊 Mathematics & scientific computing
- 🖥️ System scripting
- 🤖 AI, Machine Learning, and Data Science

Python is widely adopted because of its rich ecosystem.

## What Python Can Do

- Build server-side web applications
- Automate workflows and pipelines
- Connect to databases and handle files
- Process big data and complex math
- Support rapid prototyping → production systems


## Why Python?

- 🧩 Cross-platform (Windows, Mac, Linux, Raspberry Pi)
- 📖 English-like, easy-to-read syntax
- ✂️ Fewer lines of code
- ⚡ Interpreted → quick feedback
- 🔀 Supports multiple programming styles:
  - Procedural
  - Object-Oriented
  - Functional

## Interpreted Language

- Python executes code **line by line**
- No separate compilation step
- Errors are identified at runtime

This enables faster experimentation and debugging.

##  Writing and Running Python

Python code can be written using:
- 🗒️ Text editors
- 💻 IDEs (PyCharm, VS Code, Thonny)
- 📓 Jupyter Notebooks for experimentation

Notebooks are widely used in data science and ML workflows.
