```python
# need of file handling in python
# in general we store data in variable which is in memory but if we want to store data permanently we need to use file handling.
# file handling is a way to store data permanently in a file and we can read data from the file whenever we want.
# in python we have built-in functions to handle files.
```

```python
age = input("enter your age: ")
print("your age is: ", age)

f = open("data.txt", "w")  # open a file in write mode
f.write(age)  # write the age to the file
f.close()  # close the file

print("data has been written to the file")
```

```python
f = open("data.txt", "r")  # open the file in read mode
data = f.read()  # read the data from the file
print("data read from the file: ", data)
f.close()  # close the file
```

```python
# what is file 
# files are maned location on the disk to store information.
# they are used to store data permanently and can be accessed later. (non volatile memory)

# types of files
# 1. text files: they are used to store data in human readable format. (txt, csv, json, xml)
# 2. binary files: they are used to store data in machine readable format. (exe, dll, jpg, png)

# What is file handling 
# 1 opening a file
# 2 performing operations on the file (read, write, append)
# 3 closing the file

```

```python
# openinf a file
# open() function is used to open a file in python.
# syntax: open(file_name, mode)

# file_name: name of the file or path to be opened/accessed
# mode: mode in which the file is to be opened. (r, w, a, x, b, t)
# r: read mode (default mode) - opens the file for reading. if the file does not exist it will throw an error.
# w: write mode - opens the file for writing. if the file does not exist it will create a new file. if the file already exists it will overwrite the existing file.
# a: append mode - opens the file for writing. if the file does not exist it will create a new file. if the file already exists it will append the new data to the existing file.
# x: exclusive creation mode - opens the file for writing. if the file does not exist it will create a new file. if the file already exists it will throw an error.
# b: binary mode - opens the file in binary mode. it is used to read or write binary files.
# f - file handler, file pointer, file object - it is a variable that is used to access the file and perform operations on it.
```

```python
# opening a file
f = open("data.txt", "r")  # open a file in read mode 
if f:
    print("file opened successfully")
else:
    print("file not found")

print("file name: ", f.name)  # print the name of the file
print("file mode: ", f.mode)  # print the mode in which the file is opened
print("file closed: ", f.closed)  # print whether the file is closed or not
print("file encoding: ", f.encoding)  # print the encoding of the file
print("content of the file: ", f.read())  # read the content of the file and print it

# closing the file
f.close()
print("file closed: ", f.closed)  # print whether the file is closed or not
```
