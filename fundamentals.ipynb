## Casting (Type Conversion)

- Casting means converting a value from one data type to another.

- Python gives us simple built‑in functions to convert between types like int, float, string, etc.

This is extremely important in ML/AI workflows because:

- User input comes as strings, but models need numbers
- Data from files often needs type correction
- Converting between numeric types helps avoid errors

### Implicit Casting

- Python automatically converts smaller → larger compatible types.
- 😄 Python upgrades the value to avoid losing information.

```python

x = 5      # int
y = 2.0    # float
z = x + y  # int + float → float
print(z)   # 7.0
print(type(z))  # float
```

### Explicit Casting

- You convert the type yourself using built‑in functions:

- int() → convert to integer

- float() → convert to decimal

- str() → convert to string

```python
# string to integer and float
x = "100"
print(x, type(x))  # 100 <class 'str'>

y = int(x)
z = float(x)
print(y, type(y))      # 100 <class 'int'>
print(z, type(z))      # 100.0 <class 'float'
```

```python
# integer to string and float

x = 250
print(x, type(x))      # 250 <class 'int'>

y = str(x)
z = float(x)
print(y, type(y))      # 250 <class 'str'>
print(z, type(z))      # 250.0 <class 'float'>
```

```python
# float to string and integer

x = 99.99
print(x, type(x))      # 99.99 <class 'float'>

y = str(x)
z = int(x)
print(y, type(y))      # 99.99 <class 'str'>
print(z, type(z))      # 99 <class 'int'>
```

## Boolean

- Booleans in Python represent one of two values:
    - True
    - False

- They are used heavily in conditions, comparisons, loops, and decision‑making.

- The built‑in bool() function lets you check if a value is True or False based on Python’s rules.

```python
# Evaluate Values and Variables Using bool()
print(bool(10))      # True
print(bool(0))       # False
print(bool("Hi"))    # True
print(bool(""))      # False

```

```python
# Boolean Values from Expressions

print(5 > 2)        # True
print(10 == 3)      # False
print(7 <= 7)       # True
```

```python
# Booleans From Variables

x = 20
y = 15

print(x > y)   # True
print(bool(x)) # True because x is non‑zero
```

```python
# Booleans in Conditional Logic
if True:
    print("This will always run!")

if bool("data"):
    print("This runs because string is not empty")
```

## Operators

- These operators help you perform calculations, comparisons, logic building, identity checks, membership testing, and low-level bit manipulation.


### 1️⃣ Arithmetic Operators



```python

a, b = 15, 4

print("a + b =", a + b)     # Addition -> 19
print("a - b =", a - b)     # Subtraction -> 11
print("a * b =", a * b)     # Multiplication -> 60
print("a / b =", a / b)     # True division -> 3.75 (float)
print("a // b =", a // b)   # Floor division -> 3
print("a % b =", a % b)     # Modulus -> 3  --> Remainder of 15 divided by 4
print("a ** b =", a ** b)   # Exponent -> 15**4 = 50625

```

### 2️⃣ Assignment Operators



```python
x = 10          # Simple assignment
print("Initial x:", x)

x += 5          # x = x + 5  -> 15
print("After x += 5:", x)

x -= 3          # x = x - 3  -> 12
print("After x -= 3:", x)

x *= 2          # x = x * 2  -> 24
print("After x *= 2:", x)

x /= 4          # x = x / 4  -> 6.0 (note: becomes float)
print("After x /= 4:", x)

x %= 5          # x = x % 5  -> 1.0
print("After x %= 5:", x)

x **= 3         # x = x ** 3 -> 1.0
print("After x **= 3:", x)

x //= 2         # x = x // 2 -> 0.0
print("Final x:", x)


```

### 3️⃣ Comparison Operators

- Used to compare values — return True / False.



```python
m, n = 7, 7

print("m == n:", m == n)    # Equal      -> True
print("m != n:", m != n)    # Not equal  -> False
print("m >  n:", m > n)     # Greater    -> False
print("m <  n:", m < n)     # Less       -> False
print("m >= n:", m >= n)    # Greater or equal -> True
print("m <= n:", m <= n)    # Less or equal    -> True
```

### 4️⃣ Logical Operators

- Combine conditions



```python

is_ds_course = True
has_prereq   = False

print("and :", is_ds_course and has_prereq)  # True only if both True -> False
print("or  :", is_ds_course or has_prereq)   # True if any True       -> True
print("not :", not is_ds_course)             # Negates                -> False

# Common pattern with comparisons
if score >= 60 and score < 85:
    print("Pass with B-range")       # True
elif score >= 85:
    print("Pass with A-range")       # True
```

### 5️⃣ Identity Operators

- Check if two variables refer to the same object in memory.



```python

a = [1, 2, 3]
b = a              # b refers to the same list object
c = [1, 2, 3]      # c is a different list with same contents

print("a is b    :", a is b)         # True (same object)
print("a is c    :", a is c)         # False (different objects)
print("a == c    :", a == c)         # True (same values)
print("a is not c:", a is not c)     # True
print("a is not b:", a is not b)     # False
```

### 6️⃣ Membership Operators

- Check if a value exists in a sequence.



```python

letters = ["a", "b", "c"]
text = "machine learning"

print("'b' in letters     :", "b" in letters)          # True
print("'z' not in letters :", "z" not in letters)      # True
print("'learn' in text    :", "learn" in text)         # True
print("'AI' in text       :", "AI" in text)            # False (case-sensitive)
```

### 7️⃣ Bitwise Operators

- Operate on binary values - used in low‑level programming.



```python
a = 7
b = 3

print(a & b)   # AND 0111 & 0011 = 0011 -> 3
print(a | b)   # OR  0111 | 0011 = 0111 -> 7
print(a ^ b)   # XOR 0111 ^ 0011 = 0100 -> 4
print(a << 1)  # Left Shift 0111 << 1 = 1110 -> 14
print(a >> 1)  # Right Shift 0111 >> 1 = 0011 -> 3

```

```python

# Use small numbers so binary is easy to see:
p = 0b0101  # 5
q = 0b0011  # 3

print("p & q  :", bin(p & q),  p & q)   # AND  -> 0b1        1
print("p | q  :", bin(p | q),  p | q)   # OR   -> 0b111      7
print("p ^ q  :", bin(p ^ q),  p ^ q)   # XOR  -> 0b110      6
print("~p     :", bin(~p),     ~p)      # NOT  -> ... two's complement
print("p << 1 :", bin(p << 1), p << 1)  # Left shift  -> 0b1010    10
print("p >> 1 :", bin(p >> 1), p >> 1)  # Right shift -> 0b10      2
```

```python
# converting integer to binary  using bin() function

num = 42
binary_representation = bin(num)
binary_representation_1 = bin(num)[2:]
print(f"Binary representation of {num} is {binary_representation}")
print(f"Binary representation of {num} is {binary_representation_1}")

# coverting binary to integer using int() function

binary_str = "0b101010"
binary_str_1 = "101010"
integer_value_1 = int(binary_str_1, 2)
integer_value = int(binary_str, 2)
print(f"Integer value of binary {binary_str} is {integer_value}")
print(f"Integer value of binary {binary_str_1} is {integer_value_1}")
```
