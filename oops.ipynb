### Types of Constructor

#### Parameterized Constructor

```python
# Parameterized Constructor

class Student:
    def __init__(self, name, age):
        print("Constructor with parameters called")
        print("memory reference for object", self)
        self.name = name                                        # attribute assignment
        self.age = age

    def display(self):                                          # methods
        print(f"hello i am {self.name} and my age is {self.age}")

stu1 = Student("Alice", 20)
stu2 = Student("Bob", 22)
```

##### Actual flow !

    1)Memory is allocated for the new object.

    2)The memory reference (address) of that object is returned by Python.

    3)This memory reference is automatically passed as the first argument (self) to the constructor.
    
    4)The constructor initializes/creates attributes inside that memory block using the reference.

### Types of Variables & Methods

#### Instance Variables 

- variables that are unique to each instance of a class or object
- seperate memory is allocated for each instance variable
- values of instance variables can be different for different objects
- modification in one object does not affect other objects
- Created inside the constructor `__init__` or assigned using self


```python
class Student:
    ''' This is Student class '''
    def __init__(self, name, age):
        self.name = name     # instance variable for student's name
        self.age = age       # instance variable for student's age

stu1 = Student("Alice", 20)
stu2 = Student("Bob", 22)
print(stu1.__dict__)  
print(stu2.__dict__)  

# modification in stu1 does not affect stu2
stu1.name = "Charlie"
print(stu1.__dict__) 
print(stu2.__dict__)  

```

#### Instance Methods 

- methods that operate on instance variables
- a method which is defined inside a class and is called using an object of the class.
- It takes self as the first parameter which refers to the current instance of the class.

```python
class Student:
    ''' This is Student class '''
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):  # instance method
        print(f"hello i am {self.name} and my age is {self.age}")

stu1 = Student("Alice", 20)
stu1.display()  # calling instance method using object

```

##### getter method  - to get the value of an attribute 

##### setter method - to set the value of an attribute


- these method are called mutator and accessor methods respectively
- getter and setter methods are used to access and modify private attributes of a class
- private attributes are those attributes which are not accessible outside the class and are defined with double underscore `__` before the attribute name


```python
class Student:
    def __init__(self, name, age):
        self.__name = name   # private attribute
        self.__age = age     # private attribute

    def get_name(self):      # getter method for name
        return self.__name

    def set_name(self, name):  # setter method for name
        self.__name = name

    def get_age(self):       # getter method for age
        return self.__age

    def set_age(self, age):   # setter method for age
        self.__age = age

stu1 = Student("Alice", 20)
print(stu1.get_name())  # accessing private attribute using getter method
print(stu1.get_age())   # accessing private attribute using getter method

stu1.set_name("Charlie")  # modifying private attribute using setter method
stu1.set_age(21)         # modifying private attribute using setter method

print(stu1.get_name())  # accessing modified private attribute using getter method
print(stu1.get_age())   # accessing modified private attribute using getter method

#print(stu1.__name)  # trying to access private attribute directly will raise an error
```

#### Class method 

- methods that operate on class variables
- first argument is class referecne
- made using decorator @classmethod (built-in decorator)
- There 2 ways to modify class variable - using class name and using class method. 
- It is recommended to use class method to modify class variable as it provides better encapsulation and maintainability.

```python
class Student:
    school_name = "ABC High School"   # class variable

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):               # instance method
        print(f"hello i am {self.name} and my age is {self.age}")

    @classmethod
    def change_school(cls, new_school):   # class method to modify class variable
        cls.school_name = new_school

stu1 = Student("Alice", 20)
stu2 = Student("Bob", 22)

print(stu1.__dict__)
print(stu2.__dict__)
print('########################')

print(stu1.school_name)
print(stu2.school_name)
print(Student.school_name)                   # accessing class variable using class name 
print('########################')

# modifying class variable using class method
Student.change_school("XYZ High School")      
print(Student.school_name)
print(stu1.school_name)
print(stu2.school_name)

```

#### Static method

- Methods which performs operations on external data.
- It can also perform operations on class data.
- No need to pass object or class reference.
- created using decorator @staticmethod


```python
# example of static method

class Bank:
    bank_name = 'SBI'
    rate_of_interest = 5.5

    @staticmethod
    def simple_intrest(principal, time):
        return (principal * Bank.rate_of_interest * time) / 100

princpal_amount = 10000  # external data which is not related to class or instance
time_period = 2  # external data which is not related to class or instance

# calling static method using class name
interest = Bank.simple_intrest(princpal_amount, time_period)
print(f"Simple interest is: {interest}")

```

## Inheritance

### What ? 

- Deriving a new class from an existing class so that the new class inherits all members (attributes + methods) of the existing class is called inheritance.

- Old class :- Parent class, Base class, Existing class, Super class
- New class :- Child class, Sub class, Derived class

  

### Why ?

- for code resuability 
- having relation among classes



```python
class Employee:
    bonus = 2000
    def display(self):
        print("This is Employee class")

class Manager(Employee):  # Manager class is inheriting Employee class
    bonus1 = 5000  
    def show(self):
        print("This is Manager class")

e1 = Employee()
m1 = Manager()

e1.display()
m1.show()
m1.display() # Manager class can access display method of Employee class because of inheritance
print(e1.bonus)  # accessing class variable of Employee class using Employee object
print(m1.bonus)  # accessing class variable of Employee class using Manager object
print(m1.bonus1)  # accessing class variable of Manager class using Manager object

```

### super() function 

- Using super() function, we can access parent class properties.
- This function returns a temporary object which contains reference to parent class.
- It makes inheritance more manageable and extensible.

```python
class Computer(object): # bydefaut inherits from object class
    def __init__(self):
        print("This is Computer class constructor")
        self.ram = '8gb'
        self.storage = '512gb'

class Mobile(Computer):  # Mobile class is inheriting Computer class
    def __init__(self):
        print("This is Mobile class constructor")
        self.model = 'Iphone X'

Apple=Mobile()
print(Apple.__dict__)
```

```python
class Computer(object): # bydefaut inherits from object class
    def __init__(self):
        print("This is Computer class constructor")
        self.ram = '8gb'
        self.storage = '512gb'

class Mobile(Computer):  # Mobile class is inheriting Computer class
    def __init__(self):
        print("This is Mobile class constructor")
        super().__init__()  # calling parent class constructor to access its properties
        self.model = 'Iphone X'

Apple=Mobile()
print(Apple.__dict__)
 
```

### Types of Inheritance

#### Single Inheritance 
- One parent and one chile class



#### Multilevel Inheritance
- Parent class and  child class furthur inherited into new class forming multiple levels.




```python
# single level 
class Parent:
    def __init__(self):
        print("This is Parent class constructor")
        self.parent_attr = "I am parent attribute"

class Child(Parent):  # Child class is inheriting Parent class
    def __init__(self):
        print("This is Child class constructor")
        super().__init__()  # calling parent class constructor to access its properties
        self.child_attr = "I am child attribute"

c = Child()
print(c.__dict__)
```

```python
# multi level 

class GrandParent:
    def __init__(self):
        print("This is GrandParent class constructor")
        self.house = '1 BHK'

    def display(self):
        print(f"I am living in {self.house}")

class Father(GrandParent):  # Father class is inheriting GrandParent class
    def __init__(self):
        print("This is Father class constructor")
        super().__init__()  # calling parent class constructor to access its properties
        self.money = 100000

    def show(self):
        print(f"I am living in {self.house} and i have {self.money} money")

class Son(Father):  # Son class is inheriting Father class
    def __init__(self):
        print("This is Son class constructor")
        super().__init__()  # calling parent class constructor to access its properties
        self.vehicle = "BMW"

    def info(self):
        print(f"I am living in {self.house} and i have {self.money} money and i have {self.vehicle} vehicle")

s = Son()
print(s.__dict__)
s.info()
print("#########################")

f = Father()
print(f.__dict__)
f.show()
print("#########################")

g = GrandParent()
print(g.__dict__)
g.display()

print("#########################")
s.show() # calling show method of Father class using Son class object
s.display() # calling display method of GrandParent class using Son class object
```

```python
# multi level 

class GrandParent:
    def __init__(self):
        print("This is GrandParent class constructor")
        self.house = '1 BHK'

    def display(self):
        print(f"I am living in {self.house}")

class Father(GrandParent):  # Father class is inheriting GrandParent class
    pass

class Son(Father):  # Son class is inheriting Father class
    def __init__(self):
        print("This is Son class constructor")
        super().__init__()  # calling parent class constructor to access its properties
        self.vehicle = "BMW"

    def info(self):
        print(f"I am living in {self.house} and i have {self.money} money and i have {self.vehicle} vehicle")

s = Son()
print(s.__dict__)
print("#########################")

f = Father()
print(f.__dict__)
print("#########################")

g = GrandParent()
print(g.__dict__)
g.display()
```

#### Hierarchical Inheritance
- One parent and muliple child classes



```python
class Person:
    def __init__(self,nm, ag):
        self.name = nm
        self.age = ag
    
    def display(self):
        print(f"Hello i am {self.name} and my age is {self.age}")


class Employee(Person):
    def __init__(self, nm, ag, sal):
        super().__init__(nm, ag)  # calling parent class constructor
        self.salary = sal

    def show(self):
        print(f"Hello i am {self.name} and my age is {self.age} and my salary is {self.salary}")
    
class Student(Person):
    def __init__(self, nm, ag, marks):
        super().__init__(nm, ag)  # calling parent class constructor
        self.marks = marks

    def info(self):
        print(f"Hello i am {self.name} and my age is {self.age} and my marks are {self.marks}")

e = Employee("Alice", 30, 50000)
s = Student("Bob", 20, 85)
e.show()
s.info()

print("#########################")
print(e.__dict__)
print(s.__dict__)

# s.show() --> gives attribute error
# e.info() --> gives attribute error because show method is only defined in Employee class and info method is only defined in Student class


```

## Encapsulation 

- Wrapping up data and methods working on data together in a single unit (i.e class) is called encapsulation 

- Data Hiding & Security
- Flexibility & Maintenance

### Access Modifiers

- Generally, we restrict data access outside the class in encapsulation.

- Encapsulation can be achieved by declaring the data members and methods of a class as private.

- Three access specifiers:- public, private, protected


```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def display(self):
        print(f"Hello i am {self.name} and my salary is {self.salary}")


e1 = Employee("Alice", 50000)
print(e1.__dict__)
print(e1.name)   # now we need to restrict this access , and this encapsulation or data hiding.
e1.display()
```

```python
# example of encapulation or data hiding

class Finance:
    def __init__(self, revenue):
        self.__revenue= revenue  # private attribute

    def get_revenue(self):      # getter method for revenue
        return self.__revenue

    def set_revenue(self, revenue):  # setter method for revenue
        self.__revenue = revenue

f = Finance(1000000)
print(f.__dict__)
#print(f.revenue)   # attribute error - restricted
print(f.get_revenue())  # accessing private attribute using getter method

f.set_revenue(2000000)  # modifying private attribute using setter method
print(f.get_revenue())  # accessing modified private attribute using getter method


```

### Name Mangling

- so can we access private data using name mangling ? Yes ( directly cant access but indirectly we can using name mangling)

```python
class Finance:
    def __init__(self, revenue):
        self.__revenue= revenue  # private attribute

    def get_revenue(self):      # getter method for revenue
        return self.__revenue

    def set_revenue(self, revenue):  # setter method for revenue
        self.__revenue = revenue

f = Finance(1000000)
print(f.__dict__)


#print(f.revenue)   # attribute error - restricted

print(f.get_revenue())  # accessing private attribute using getter method

print(f._Finance__revenue)  # name mangling to access private attribute directly (not recommended)
```

### Polymorphism with Inheritance

```python
# example Polymorphism with Inheritance

class Vehicle:
    def __init__(self,name,color,price):
        self.name = name
        self.color = color
        self.price = price

    def get_info(self):
        print(f"Vehicle Name: {self.name}, Color: {self.color}, Price: {self.price}")

    def max_speed(self):
        print("Max speed of vehicle is 100 km/h")

    def gear_type(self):
        print("Vehicle has 6 gear")

class Car(Vehicle):
    def max_speed(self):
        print("Max speed of car is 200 km/h")

    def gear_type(self):
        print("Car has 5 gear")

v1 = Vehicle("Generic Vehicle", "White", 5000)
c1 = Car("Sedan", "Black", 20000)

v1.get_info()
c1.get_info()

print(v1.__dict__)
print(c1.__dict__)

v1.max_speed()  # calls Vehicle class method
v1.gear_type()  # calls Vehicle class method

c1.max_speed()  # calls Car class method (overridden)
c1.gear_type()  # calls Car class method (overridden)
 
  
```

### overriding built-in functions

- magic methods work by allowing developers to override built-in functions to change their functionality.

- These methods, which start and end with double underscores (e.g.,` __len__`), are called automatically by Python when a corresponding built-in function is used on an object.

- To overload an operator in Python, you use the Dunder (Double Underscore)



```python
class Cart:
    pass

c1 = Cart()
print(c1)
```

```python
class Cart:
    def __str__(self):
        return "This is a Cart object"

c1 = Cart()
print(c1)
```

```python
class Cart:
    def __init__(self,basket1,basket2,basket3):
        self.clothes = basket1
        self.electronics = basket2
        self.others = basket3

c1 = Cart(["Shirt", "Pants"], ["Phone", "Laptop"], ["Book", "Toy"])
print(c1.__dict__)

#print(len(c1)) # this will give error because len() function does not know how to calculate length of Cart object
    
```

```python
class Cart:
    def __init__(self,basket1,basket2,basket3):
        self.clothes = basket1
        self.electronics = basket2
        self.others = basket3

    def __len__(self):
            return len(self.clothes) + len(self.electronics) + len(self.others)

c1 = Cart(["Shirt", "Pants"], ["Phone", "Laptop"], ["Book", "Toy"])
print(c1.__dict__)

print(len(c1)) # now len() function can calculate length of Cart object because we have defined __len__ method in Cart class



# How it worked ? 
# 1 when we called len(c1) , it checked c1 is of which class and it found c1 is of Cart class
# 2 then it checked Cart class has any magic method defined for len() function and it found __len__ method is defined in Cart class
# 3 then it called __len__ method of Cart class and it calculated the length of Cart object and returned the value
```

```python
class BMW:
    def fuel_type(self):
        print("BMW runs on Diesel")

    def max_speed(self):
        print("BMW has max speed of 250 km/h")

class Ferrari:
    def fuel_type(self):
        print("Ferrari runs on petrol")

    def max_speed(self):
        print("Ferrari has max speed of 300 km/h")

def car_details(car):
    car.fuel_type()
    car.max_speed()

bmw = BMW()
ferrari = Ferrari()

car_details(bmw)
print("#########################")
car_details(ferrari)
```

### Operator Overloading

- When same operator behaves differently depending on values.

- You can assign a new meaning to operators also and you can extend functionality of operators.

- You can change default behaviour of operator using over-riding. 

```python
num1 = 10
num2 = 20 
print(num1 + num2)  # addition of numbers
print(num1.__add__(num2))  # using magic method __add__
print(int.__add__(num1, num2))  # using int class __add__ method directly


# how code works ? 
# 1- check datatype of left operand (num1) and it found num1 is of int type
# 2 - go to that class
# 3 - call _add_() function of int class and pass num1 and num2 as arguments
```

```python
print(dir(int))  # to see all the methods of int class and we can see __add__ method is defined in int class
```

```python
num1 = 'hello'
num2 = 'world'

print(num1 + num2)  # concatenation of strings
print(num1.__add__(num2))  # using magic method __add__
print(str.__add__(num1, num2))  # using str class __add__ method directly

print(dir(str))  # to see all the methods of str class and we can see __add__ method is defined in str class
```

```python
class Score:
    def __init__(self, value):
        self.value = value

    # This overloads the '+' operator
    def __add__(self, other):
        # We return a new Score object
        return Score(self.value + other.value)

    # This allows us to see the number when we print it
    def __str__(self):
        return str(self.value)

# Using it:
player1 = Score(100)
player2 = Score(50)

# The '+' magically calls the __add__ method
total = player1 + player2
print(type(total))
print(total.__dict__)

print(total)  # Output: 150
```

### Method Overloading

- When a class contains two or methods with same name but different behaviour.

- In Python, if you define two methods with the same name, the second one overwrites the first one entirely


```python
class Calculator:
    def add(self, a, b):
        return a + b

    # This second definition REPLACES the first one!
    def add(self, a, b, c):
        return a + b + c

calc = Calculator()
# print(calc.add(5, 5))  <-- This would now throw an ERROR!
print(calc.add(5, 5, 10)) # This works.
```

```python
# method overloading acheiving by some modification

class Area:
    def area(self, a, b=None):
        if b is None:
            return a * a  # Area of square
        else:
            return a * b  # Area of rectangle
        
a = Area()
print(a.area(5))      # Area of square with side 5
print(a.area(5, 10))  # Area of rectangle with length 5 and
```

## Nested Class

- The class which is declared inside another class.

- When one class obj cannot exist w/o another class object.




```python
class Outer:
    def __init__(self):
        print("This is Outer class constructor  ")

    def display(self):
        print("This is display method of Outer class")

    class Inner:
        def __init__(self):
            print("This is Inner class constructor")

        def show(self):
            print("This is show method of Inner class")

o = Outer()  # creating object of Outer class
o.display()  # calling method of Outer class

i = Outer.Inner()  # creating object of Inner class
i.show()  # calling method of Inner class
```

## Destructor

- A special method which destroys objects and releases resources tied to objects.

- Destructor is called automatically when object is destroyed.



- Below two conditions when destructor is called 
 - Reference counting reaches to 0
 - When variable goes out of scope

- Note:- In Python, The special method `__del__`() is used to define a destructor.
 



Disadvantage 

- Cicurlar referencing




```python
# example of destructor

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        print(f"Employee {self.name} with salary {self.salary} is created")

    def display(self):
        print(f"Hello i am {self.name} and my salary is {self.salary}")
    
    # defining destructor method
    def __del__(self):
        print("destructor is called")
        print(f"Employee {self.name} with salary {self.salary} is deleted")


e1 = Employee("Alice", 50000)
print(e1.__dict__)
e1.display()  # calling method of Employee class

del e1  # deleting object e1 which will call destructor method
```

```python
import time

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Creating {self.name}")

    def __del__(self):
        print(f"--- DESTRUCTOR CALLED: {self.name} is now deleted! ---")

# 1. Create the object and point two variables at it
obj1 = MyObject("TestObject")
obj2 = obj1  # Now both obj1 and obj2 point to the SAME object (Ref Count = 2)

print("\n--- Action: Deleting obj1 ---")
del obj1  # We deleted the name 'obj1', but the object still exists in 'obj2'
time.sleep(1) # Just to show nothing happens immediately
print("Notice: The destructor was NOT called yet.")

print("\n--- Action: Deleting obj2 ---")
del obj2  # Now the Ref Count hits ZERO.
print("This line will print AFTER the destructor.")
```

## Class Decorators

- Function Decorators
- Class Decorators

```python
# fuction decorator

def decor(func):
    def inner(a, b):
        print("Before calling the function")
        result = func(a, b)  # calling the original function
        print("After calling the function")
        output = pow(result, 2)
        return output
    return inner

@decor
def add(a,b):
    return a + b

print(add(5, 10))  # this will call the inner function of decor which will call the add function and then it will return the result after modification

```

```python
class PowerDecorator:
    def __init__(self, func):
        # Store the original function
        self.func = func

    def __call__(self, a, b): # This method is called when the decorated function is called
        print("Before calling the function")
        result = self.func(a, b)
        print("After calling the function")
        return pow(result, 2)

@PowerDecorator 
def add(a, b):
    return a + b

print(add(5, 10))  # Output: 225
```

```python
# @property decorator simple example

class Circle:
    def __init__(self, radius):
        self._radius = radius  # Using a protected attribute to store the radius

    @property
    def radius(self):
        """The radius property - getter method"""
        return self._radius

    @property
    def area(self):
        """A read-only property to calculate area of the circle"""
        import math
        return math.pi * (self._radius ** 2)
    
# Example usage:
circle = Circle(5)
print(circle.radius)  # Accessing the radius using the property
print(circle.area)    # Accessing the area using the property

```

## Abstraction 

- Only essentials details are seen, complex details are hidden





```python
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    def color(self):
        print("Car color is red")

class BMW(Car):
    def start_engine(self):
        print("BMW engine started")

class Audi(Car):
    def start_engine(self):
        print("Audi engine started")

class Tesla(Car):
    def start_engine(self):
        print("Tesla engine started")

# c1 = Car() # This will raise an error because we cannot create object of abstract class

b1 = BMW()
a1 = Audi()
t1 = Tesla()

b1.start_engine()  # calling method of BMW class
a1.start_engine()  # calling method of Audi class
t1.start_engine()  # calling method of Tesla class
```

### What ?

- A class which contains one or more abstract methods and concrete methods.

- Abstract class must have at least one abstract method.

- An abstract class can be considered as a blueprint for other classes.

### Keep in mind..

- You can't instantiate abstract classes.

- Abstract class requires at least one method abstract.

- All abstract methods present in abstract class must be implemented in child classes. Else, child class becomes abstract.



#### Uses of abstraction

- When we have large code and many functionalities.
Your abstract class acts like an API for other subclasses.

- Used when a third party is going to provide implementations.

- When we have different implementations for different objects for the same component.
For creating interfaces.


```python
from abc import ABC, abstractmethod

# 1. The Abstract Class (The Blueprint/Contract)
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass  # No code here! Just a requirement.

# 2. The Concrete Class (The Implementation)
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started with a key!"

class Tesla(Vehicle):
    def start_engine(self):
        return "Tesla started with a mobile app!"

# Using the abstraction
my_car = Car()
my_tesla = Tesla()

print(my_car.start_engine())
print(my_tesla.start_engine())
```

```python
from abc import ABC, abstractmethod

# 1. THE ADT (The Abstract Definition)
# This says: "Any List must have these 3 behaviors"
class ListADT(ABC):
    @abstractmethod
    def append(self, item):
        """Add an item to the end of the list"""
        pass

    @abstractmethod
    def get_at(self, index):
        """Retrieve an item by its position"""
        pass

    @abstractmethod
    def display(self):
        """Show all items in the list"""
        pass

# 2. THE IMPLEMENTATION (The Concrete Class)
# This is where we actually write the code to make it work.
class MyCustomList(ListADT):
    def __init__(self):
        # We are using a private list to store our data
        self.__internal_data = []

    def append(self, item):
        self.__internal_data.append(item)
        print(f"Added {item} to the list.")

    def get_at(self, index):
        if 0 <= index < len(self.__internal_data):
            return self.__internal_data[index]
        return "Error: Index out of bounds"

    def display(self):
        return f"Current List: {self.__internal_data}"

# 3. USING THE ADT
# The user only cares about the methods defined in the ADT.
my_list = MyCustomList()
my_list.append("Python")
my_list.append("Java")
print(my_list.get_at(0))
print(my_list.display())
```
